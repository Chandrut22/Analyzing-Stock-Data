```python
import pandas as pd
```

```python
ad_clicks = pd.read_csv('https://raw.githubusercontent.com/ChanBotCoder/A-B-Testing-for-ShoeFly.com/main/ad_clicks.csv')
```

```python
ad_clicks.head()
```

```python
ad_clicks.groupby('utm_source').user_id.count().reset_index()
```

```python
ad_clicks['is_click'] = ad_clicks.ad_click_timestamp.isnull()
```

```python
clicks_by_source = ad_clicks.groupby(['utm_source','is_click']).user_id.count().reset_index()
clicks_by_source
```

```python
clicks_pivot = clicks_by_source.pivot(index = 'utm_source', columns = 'is_click' , values = 'user_id').reset_index()
clicks_pivot
```

```python
clicks_pivot['percent_clicked'] = clicks_pivot[True] / (clicks_pivot[True] + clicks_pivot[False])
```

```python
print(ad_clicks.groupby('experimental_group').user_id.count().reset_index())
```

```python
print(ad_clicks.groupby(['experimental_group','is_click']).user_id.count().reset_index().pivot(columns = 'is_click' , index = 'experimental_group',values = 'user_id').reset_index())
```

```python
a_clicks = ad_clicks[ad_clicks.experimental_group == 'A']
a_clicks.head()
```

```python
b_clicks = ad_clicks[ad_clicks.experimental_group == 'B']
b_clicks.head()
```

```python
a_clicks_pivot = a_clicks.groupby(['is_click','day']).user_id.count().reset_index().pivot(index = 'day',columns = 'is_click',values = 'user_id').reset_index()
a_clicks_pivot
```

```python
a_clicks_pivot['percent_clicked'] = a_clicks_pivot[True] / (a_clicks_pivot[True] + a_clicks_pivot[False])
```
